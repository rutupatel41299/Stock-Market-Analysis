```
# Import required libraries
import pandas as pd
import yahoo_fin.stock_info as si
from yahoo_fin.stock_info import get_data
import warnings
from azure.storage.blob import BlobServiceClient
import pyarrow as pa
import pyarrow.parquet as pq
from delta import *

# Get S&P 500 tickers
ticker_df = pd.DataFrame(si.tickers_sp500())
ticker_list = ticker_df[0].values.tolist()

# Remove specific tickers
stringToRemove = ['KVUE', 'VLTO']
ticker_list = [item for item in ticker_list if item not in stringToRemove]

# Initialize an empty DataFrame to store the data
output_df = pd.DataFrame()
warnings.filterwarnings("ignore")

# Loop through tickers and fetch data
for ticker in ticker_list:
    print("Fetching data for ticker:", ticker)
    data = get_data(ticker, start_date="01/01/2000", end_date="12/31/2022", index_as_date=True, interval="1d")
    output_df = output_df.append(data)

# Display the fetched data
display(output_df)
```

```
# by default date is index column, so we added another column named date
output_df['date'] = pd.to_datetime(output_df.index).date
output_df
```
